# Pattern programs in  Day-1

```python
n=int(input())
for i in range(n):
    for j in range(n):
        if(i==0 or i==n-1 or j==0 or j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
n=int(input())
for i in range(1,n+1):
    for j in range(1,2*n):
        if i==n or i+j==n+1 or j-i==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
